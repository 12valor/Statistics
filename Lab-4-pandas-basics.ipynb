## Descriptive Analytics <hr style="border:2.5px solid rgba(255, 0, 0, 1)"></hr>

```python
# Jaguar vs Panther
resistance = np.array([
    [10.6,9.1,9.3,9.8,10.5,10.4,9.5,11,10.4,3,9.8],
    [10.1,11,9.1,20,9.2,10.8,9.9,9.2,9.1,9.1,np.nan]
]).T
resistance
```

```python
df = pd.DataFrame(resistance,columns=['Jaguar','Panther'])
df
```

```python
df['Panther']
```

```python
# Summary of DataFrame
df.info()
```

```python
df.isna()
```

```python
df.dropna()
```

```python
df.fillna(100)
```

```python
df['Panther'].fillna(df['Panther'].mean())
```

```python
df
```

```python
# Summary of statistics
df.describe().round(2)
```

```python
# Median
jag_median = df['Jaguar'].median()
pan_median = df['Panther'].median()

print(f'Jaguar Median = {jag_median:.2f} ohms')
print(f'Panther Median = {pan_median:.2f} ohms')
```

```python
# Mean
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

print(f'Jaguar Mean = {jag_mean:.2f} ohms')
print(f'Panther Mean = {pan_mean:.2f} ohms')
```

```python
# Mode
jag_mode = df['Jaguar'].mode()
pan_mode = df['Panther'].mode()
```

```python
#pd.DataFrame(jag_mode)
print('Jaguar Mode:')

for i in jag_mode:
    print(i)
```

```python
#pd.DataFrame(jag_mode)
print('Panther Mode:')

for i in pan_mode:
    print(i)
```

```python
# Jaguar IQR
jag_Q1 = df['Jaguar'].quantile(0.25)
jag_Q3 = df['Jaguar'].quantile(0.75)

print(f'Jaguar IQR {jag_Q3-jag_Q1:.2f} ohms')
```

```python
# Panther IQR
pan_Q1 = df['Panther'].quantile(0.25)
pan_Q3 = df['Panther'].quantile(0.75)

print(f'Panther IQR {pan_Q3-pan_Q1:.2f} ohms')
```

## Bar Chart <hr style="border:2.5px solid rgba(255, 0, 0, 1)"></hr>

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

plt.figure(figsize=(5,7))
bars = plt.bar(['Jaguar','Panther'],[jag_mean,pan_mean],color='red')

plt.title('Mean Comparison Between Machines')
plt.ylabel('Resistance (ohms)')
plt.xlabel('Machines')

# Label
plt.bar_label(bars,padding=2)
plt.margins(y=0.5)
plt.show()
```

```python
jag_median = df['Jaguar'].median()
pan_median = df['Panther'].median()

plt.figure(figsize=(5,6))
bars = plt.bar(['Jaguar','Panther'],[jag_median,pan_median],color='blue')

plt.title('Median Comparison Between Machines')
plt.ylabel('Resistance (ohms)')
plt.xlabel('Machines')

# Label
plt.bar_label(bars,padding=2)
plt.margins(y=0.5)
plt.show()
```
